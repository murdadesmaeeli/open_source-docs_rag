````python
import requests
from bs4 import BeautifulSoup

def build_nested_tree_string(tag, level=0, max_depth=10):
    """
    Recursively builds a nested string representation of the DOM subtree
    rooted at 'tag', but only up to 'max_depth' levels deep.
    Includes the element's 'class' attribute if it exists.
    """
    # If we've exceeded our depth, or the element isn't a standard tag (e.g. NavigableString), stop.
    if level > max_depth or tag.name is None:
        return ""
    
    # Construct a class string if the tag has classes
    class_str = ""
    if 'class' in tag.attrs:
        class_list = tag.get('class', [])
        if class_list:
            class_str = f" class='{ ' '.join(class_list) }'"
    
    # Build the line describing this element
    line = "  " * level + f"<{tag.name}{class_str}>"
    
    # We’ll store lines in a list, then join them with newlines
    lines = [line]
    
    # Use recursive=False so we only go one level deeper per function call
    children = tag.find_all(recursive=False)
    for child in children:
        child_repr = build_nested_tree_string(child, level + 1, max_depth)
        if child_repr:
            lines.append(child_repr)
    
    return "\n".join(lines)

def main():
    # Replace with your desired URL
    url = MAIN_URL
    
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL: {e}")
        return
    
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Find the body tag
    body_tag = soup.find("body")
    if not body_tag:
        print("No <body> tag found in the HTML.")
        return
    
    # Build the nested tree representation up to 10 levels
    tree_string = build_nested_tree_string(body_tag, level=0, max_depth=10)
    print(tree_string)

if __name__ == "__main__":
    main()
````
